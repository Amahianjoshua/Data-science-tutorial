### üßÆ Simple Linear Regression Formula

The simple linear regression model is given by:

$$
\hat{y} = b_0 + b_1 x
$$

where:

$$
\hat{y} \text{ = predicted salary}
$$

$$
x \text{ = years of experience}
$$

$$
b_0 \text{ = intercept (predicted salary when } x=0)
$$

$$
b_1 \text{ = slope (change in salary per additional year of experience)}
$$

---



```python
# importing the neccessary dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file into a Pandas DataFrame
data=pd.read_csv(r"C:\Users\joshua\Documents\Josh\@datasets\Salary_Data.csv")
data.head()
```

```python
data.shape
```

```python
# plotting the data
plt.scatter(data['YearsExperience'], data['Salary'], color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title(" Simple Linear Regression: Years of Experience Vs Salary")
plt.grid(True)
plt.show()
```

```python
# Train the Linear Regression Model
from sklearn.linear_model import LinearRegression

X=data[['YearsExperience']]
y=data['Salary']

model=LinearRegression()
model.fit(X, y)
```

```python
# Print slope and intercept
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

### Interpretation of the Linear Regression Model

The regression equation is:

$$
\hat{y} = 25,792.20 + 9,449.96 \times x
$$

where:  
- x is Work Experience (in years)  
- ≈∑ is the Predicted Salary

**Interpretation:**  
- **Intercept (b ‚âà 25,792):** This represents the estimated starting salary when work experience is zero years.  
- **Slope (m ‚âà 9,450):** This indicates that for each additional year of work experience, the salary is expected to increase by approximately 9,450 units.

This means work experience has a strong positive influence on salary, with salary increasing steadily as experience increases.


```python
# Predict salary for someone with 5 years experience
new_data=pd.DataFrame({'YearsExperience': [5]}) 
predicted_salary=model.predict(new_data)
print("The predicted salary for someone with 5 years experience is :", predicted_salary)
```

**For prediction**

```python
# predict for all the data
y_pred=model.predict(X)
```

```python
# Plot actual vs predicted 
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.title('Actual vs Predicted Salary')
plt.savefig('Actual vs Predicted Salary.png')
plt.show()
```

**üìä Scatter Plot: Relationship between Years of Experience and Salary**

- üîµ The **blue dots** represent the actual salaries from the dataset  
- üî¥ The **red line** represents the predictions made by the linear regression model

From the plot, it is clearly seen that most of the actual salaries are close to the predicted line, meaning that the model captures the trend well.  
This implies that the model has learned a strong linear relationship between work experience and salary. üíºüí∞üôÇ


```python
# Evaluation metrics
from sklearn.metrics import r2_score, mean_absolute_error
r2=r2_score(y, y_pred)
mae=mean_absolute_error(y, y_pred)

print("R¬≤ Score:", r2)
print("Mean Absolute Error:", mae)
```

**üìù Model Evaluation Summary (Simple Terms)**

- üìà The model explains **96% of the variation in salary** using just work experience. This means the predictions are highly accurate. ‚úîÔ∏è
- üíµ On average, the predicted salaries are off by only **‚Ç¶4,644**, which shows the model is quite reliable in its estimates. ü§ù
